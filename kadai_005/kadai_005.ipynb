```python
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(2, 3, tight_layout=True)

# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

# sin関数
x1 = np.linspace(0, 10, 100)
y0 = 2 + 2 * np.sin(2 * x1)

axes[0, 1].plot(x1, y0, linewidth=2.0)
axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 11), ylim=(-1, 5),
               yticks=np.arange(-1, 6))

# ヒストグラム
x２ = np.random.normal(15, 5, 2000)
axes[0, 2].hist(x２)

# 散布図
np.random.seed(3)
x0 = 4 + np.random.normal(0, 2, 60)
y１ = 4 + np.random.normal(0, 2, len(x0))
axes[1, 0].scatter(x０, y１)

plt.show()
```
